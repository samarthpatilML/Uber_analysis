# setup

```python
import io
import pandas as pd
import requests
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load the data

```python
file_name = 'uber_data.csv'
df = pd.read_csv(file_name)
```

```python
df.head()
```

# Display the first few rows of the dataframe

```python
print("First few rows of the dataframe:")
print(df.head())
```

# Inspect the data types

```python
print("\nData types:")
print(df.dtypes)
```

# Convert datetime columns to datetime format

```python
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'], errors='coerce')
df['tpep_dropoff_datetime'] = pd.to_datetime(df['tpep_dropoff_datetime'], errors='coerce')
```

# Check for any NaT (Not a Time) values that indicate invalid dates

```python
print("\nCheck for NaT values after conversion:")
print(df[['tpep_pickup_datetime', 'tpep_dropoff_datetime']].isna().sum())
```

# Drop rows with NaT values in datetime columns if necessary

```python
df = df.dropna(subset=['tpep_pickup_datetime', 'tpep_dropoff_datetime'])
```

# Extract datetime components

```python
datetime_dim = df[['tpep_pickup_datetime', 'tpep_dropoff_datetime']].reset_index(drop=True)
datetime_dim['pick_hour'] = datetime_dim['tpep_pickup_datetime'].dt.hour
datetime_dim['pick_day'] = datetime_dim['tpep_pickup_datetime'].dt.day
datetime_dim['pick_month'] = datetime_dim['tpep_pickup_datetime'].dt.month
datetime_dim['pick_year'] = datetime_dim['tpep_pickup_datetime'].dt.year
datetime_dim['pick_weekday'] = datetime_dim['tpep_pickup_datetime'].dt.weekday

datetime_dim['drop_hour'] = datetime_dim['tpep_dropoff_datetime'].dt.hour
datetime_dim['drop_day'] = datetime_dim['tpep_dropoff_datetime'].dt.day
datetime_dim['drop_month'] = datetime_dim['tpep_dropoff_datetime'].dt.month
datetime_dim['drop_year'] = datetime_dim['tpep_dropoff_datetime'].dt.year
datetime_dim['drop_weekday'] = datetime_dim['tpep_dropoff_datetime'].dt.weekday
```

# Add datetime_id

```python
datetime_dim['datetime_id'] = datetime_dim.index
```

# Reorder columns

```python
datetime_dim = datetime_dim[['datetime_id', 'tpep_pickup_datetime', 'pick_hour', 'pick_day', 'pick_month', 'pick_year', 'pick_weekday',
                             'tpep_dropoff_datetime', 'drop_hour', 'drop_day', 'drop_month', 'drop_year', 'drop_weekday']]
```

# Display the result

```python
print("\nDatetime Dimension Table:")
print(datetime_dim.head())
```

# Dictionary for mapping RatecodeID to rate code names

```python
rate_code_type = {
    1: "Standard rate",
    2: "JFK",
    3: "Newark",
    4: "Nassau or Westchester",
    5: "Negotiated fare",
    6: "Group ride"
}
```

# Creating the rate_code_dim DataFrame

```python
rate_code_dim = df[['RatecodeID']].reset_index(drop=True)
rate_code_dim['rate_code_id'] = rate_code_dim.index
rate_code_dim['rate_code_name'] = rate_code_dim['RatecodeID'].map(rate_code_type)
```

# Reordering columns

```python
rate_code_dim = rate_code_dim[['rate_code_id', 'RatecodeID', 'rate_code_name']]
```

# Display the result

```python
print("\nRate Code Dimension Table:")
print(rate_code_dim.head())
```


# Basic Exploratory Data Analysis
## Total number of rides

```python
total_rides = len(df)
print("\nTotal number of rides:", total_rides)
```

# Total number of rides by rate code

```python
rides_by_rate_code = rate_code_dim['rate_code_name'].value_counts()
print("\nTotal number of rides by rate code:")
print(rides_by_rate_code)
```

# Average trip duration in minute

```python
df['trip_duration'] = (df['tpep_dropoff_datetime'] - df['tpep_pickup_datetime']).dt.total_seconds() / 60
average_trip_duration = df['trip_duration'].mean()
print("\nAverage trip duration (minutes):", average_trip_duration)
```

# Visualization
# Histogram of trip duration

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['trip_duration'], bins=50, kde=True)
plt.title('Histogram of Trip Duration (minutes)')
plt.xlabel('Trip Duration (minutes)')
plt.ylabel('Frequency')
plt.show()
```

# Count of rides by rate code

```python
plt.figure(figsize=(10, 6))
sns.countplot(y=rate_code_dim['rate_code_name'], order=rate_code_dim['rate_code_name'].value_counts().index)
plt.title('Count of Rides by Rate Code')
plt.xlabel('Count')
plt.ylabel('Rate Code')
plt.show()
```

# Trips by pickup hour

```python
plt.figure(figsize=(10, 6))
sns.countplot(x=datetime_dim['pick_hour'], order=range(24))
plt.title('Trips by Pickup Hour')
plt.xlabel('Pickup Hour')
plt.ylabel('Count')
plt.show()
```


# Additional Visualizations
# Trips by pickup day of the week

```python
plt.figure(figsize=(10, 6))
sns.countplot(x=datetime_dim['pick_weekday'], order=range(7))
plt.title('Trips by Pickup Day of the Week')
plt.xlabel('Pickup Day of the Week')
plt.ylabel('Count')
plt.show()
```

# Trips by dropoff day of the week

```python
plt.figure(figsize=(10, 6))
sns.countplot(x=datetime_dim['drop_weekday'], order=range(7))
plt.title('Trips by Dropoff Day of the Week')
plt.xlabel('Dropoff Day of the Week')
plt.ylabel('Count')
plt.show()
```

# Trips by pickup month

```python
plt.figure(figsize=(10, 6))
sns.countplot(x=datetime_dim['pick_month'], order=range(1, 13))
plt.title('Trips by Pickup Month')
plt.xlabel('Pickup Month')
plt.ylabel('Count')
plt.show()
```

# Average trip duration by hour of the day

```python
plt.figure(figsize=(10, 6))
hourly_trip_duration = df.groupby(df['tpep_pickup_datetime'].dt.hour)['trip_duration'].mean()
sns.lineplot(x=hourly_trip_duration.index, y=hourly_trip_duration.values)
plt.title('Average Trip Duration by Hour of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Average Trip Duration (minutes)')
plt.show()
```

# Total trips over time

```python
plt.figure(figsize=(10, 6))
daily_trips = df.groupby(df['tpep_pickup_datetime'].dt.date).size()
daily_trips.plot()
plt.title('Total Trips Over Time')
plt.xlabel('Date')
plt.ylabel('Total Trips')
plt.show()
```
